```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# style for plots
sns.set_style("whitegrid")
plt.figure(figsize=(12,6))
```

```python
ticker ="AAPL"
df = yf.download(ticker,start= "2024-01-01", end="2025-08-01").copy()
```

```python
df["Daily Return %"] = df["Close"].pct_change()*100
```

```python
print(df.head())
```

```python
df["Close"].plot(title= f"{ticker} Stock Price" , color = "royalblue")
plt.ylabel("Price($)")
plt.tight_layout()
plt.show()
```

```python
sns.histplot(df["Daily Return %"].dropna(), bins =20, kde= True, color="green")
plt.title(f"{ticker} Daily Return Distribution")
plt.xlabel("Daily Return in %")
plt.tight_layout()
plt.show()
```

```python
df_sorted= df.sort_values("Daily Return %", ascending=False)
with pd.option_context('display.max_rows', None):
    display(df_sorted)
```

```python
# simple moving Average
df['SMA_20'] = df["Close"].rolling(20).mean()
df['SMA_50'] = df["Close"].rolling(50).mean()
```

```python
df[["Close", "SMA_20", "SMA_50"]].plot(title=f"{ticker} price moving average")
```

```python
df["Signal_SMA"]= 0
df.loc[df["SMA_20"] > df["SMA_50"], "Signal_SMA"] = 1
df.loc[df["SMA_20"] < df["SMA_50"], "Signal_SMA"] =-1
```

```python
plt.plot(df.index, df['Close'], label='Price', color='royalblue', alpha=0.7)
plt.scatter(
    df.index[df['Signal_SMA']== 1],
    df['Close'][df['Signal_SMA']== 1],
    label='Buy signal (SMA)',
    color='green',
    marker='^',
    s=100
)
plt.scatter(
    df.index[df['Signal_SMA']== -1],
    df['Close'][df['Signal_SMA']== -1],
    label='Sell signal (SMA)',
    color='red',
    marker='v',
    s=100
)
plt.title(f"{ticker} Trading Signals(20/50 SMA Crossover)")
plt.ylabel("Price ($)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# relative strength index as a second signal
delta = df['Close'].diff(1)
gain = delta.where(delta > 0,0)
loss = - delta.where(delta < 0,0)
avg_gain = gain.rolling(14).mean()
avg_loss = loss.rolling (14).mean()
rs = avg_gain/avg_loss
df['RSI']= 100 - (100/(1+rs))

# generate RSI signals
df['Signal_RSI']= 0
df.loc[df['RSI']< 30, 'Signal_RSI'] = 1
df.loc[df['RSI']> 70, 'Signal_RSI'] = -1

# Plot
plt.figure(figsize = (12,5))
plt.plot(df.index,df['RSI'], label='RSI (14)', color = 'purple')
plt.axhline(70, linestyle='--', color = 'red', alpha= 0.5, label= 'overbought(70)')
plt.axhline(30, linestyle='--', color = 'green', alpha= 0.5, label= 'oversold(30)')
plt.title(f'{ticker} Relative Strength Index(RSI)')
plt.legend()
plt.tight_layout()
plt.show()

# Plot RSI signals on chart
plt.figure(figsize = (12,5))
plt.plot(df.index,df['Close'], label='Price', color = 'royalblue', alpha=0.7)
plt.scatter(
    df.index[df['Signal_RSI']==1], 
    df['Close'][df['Signal_RSI']==1], 
    label='Buy signal (RSI)',
    color = 'lime',
    marker='^',
    s=100
)
plt.scatter(
    df.index[df['Signal_RSI']== -1], 
    df['Close'][df['Signal_RSI']==-1], 
    label='Sell signal (RSI)',
    color = 'darkred',
    marker='v',
    s=100
)
plt.title(f'{ticker} Trading Signals (RSI Strategy)')
plt.ylabel("Price ($)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Combined signal (SMA + RSI)
df['Signal_Combined']= 0
df.loc[(df['Signal_SMA']==1) & (df['Signal_RSI']== 1), 'Signal_Combined'] = 1
df.loc[(df['Signal_SMA']==-1) & (df['Signal_RSI']== -1), 'Signal_Combined'] = -1

#plot
plt.plot(df.index,df['Close'], label='Price', color = 'royalblue', alpha=0.7)
plt.scatter(
    df.index[df['Signal_Combined']==1], 
    df['Close'][df['Signal_Combined']==1], 
    label='Buy signal (SMA + RSI)',
    color = 'darkgreen',
    marker='^',
    s=100
)
plt.scatter(
    df.index[df['Signal_Combined']== -1], 
    df['Close'][df['Signal_Combined']==-1], 
    label='Sell signal (SMA + RSI)',
    color = 'maroon',
    marker='v',
    s=100
)
plt.title(f'{ticker} Trading Signals (SMA + RSI)')
plt.ylabel("Price ($)")
plt.legend()
plt.tight_layout()
plt.show()

```
